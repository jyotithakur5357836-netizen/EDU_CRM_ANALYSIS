```python
def get_value(value):
    if value < 40.00:
        return "Poor Administration"
    elif 40.00 <= value < 80.00:
        return "Good Administration"
    else:
       return "Need To Improve"
# Create a new column using map()
final_df['CityAdministration'] = ChargesheetingRate2022.map(get_value)


#printing result
result = final_df['CityAdministration']
print(result)
```
